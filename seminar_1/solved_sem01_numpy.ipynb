### 1.3 Как оформлять код, взятый из Интернета

```python
# https://stackoverflow.com/questions/36108824/how-to-add-elements-to-a-list-in-python

def main():
    number_of_values = int(input('Please enter number of values: '))  # int

    myList = create_list(number_of_values)  # myList = function result
    total = get_total(myList)

    print('the list is: ', myList)
    print('the total is ', total)

def get_total(value_list):
    total = 0
    for num in value_list:
        total += num
    return total

def create_list(number_of_values):
    myList = []
    for _ in range(number_of_values):  # no need to use num in loop here
        num = int(input('Please enter number: '))  # int
        myList.append(num)
    return myList

if __name__ == '__main__':  # it's better to add this line as suggested
    main()
```

### 1.5 Про команды

Полная документация: https://devpractice.ru/python-lesson-6-work-in-jupyter-notebook/

В Jupyter Notebook есть два режима работы: режим _команд_ и режим _редактирования_

_Командный_ режим нужен для того, чтобы взаимодействовать и управлять ячейками (добавлять, удалять, запускать, копировать, ...)

В режиме _редактирования_ вы меняете содержимое ячейки. 

Ячейки бывают двух основных типов, _код_ и _разметка_

### Полезные команды

(находясь в командном режиме)

- `a` - добавить пустую ячейку сверху

6. Вычисляющую первообразную заданного полинома (в качестве константы возьмите ваше любимое число). Например, если на вход поступает массив коэффициентов `array([4, 6, 0, 1])`, что соответствует полиному $4x^3 + 6x^2 + 1$, на выходе получается массив коэффициентов `array([1, 2, 0, 1, -2])`, соответствующий полиному $x^4 + 2x^3 + x - 2$. _Функции, которые могут пригодиться при решении: `.append()`_
7. Пользуясь пунктом 6, посчитайте первую производную для заданного полинома в заданной точке.

```python
# 6

# Что происходит
# [4, 6, 0, 1]
# -> [1, 2, 0, 1, -2]

# 1. Добавим константу: [4, 6, 0, 1, -2]
# 2. Степени: [4, 3, 2, 1, 0]
# 3. Корректирорвка: [4, 3, 2, 1, 1]
# 4. Делим [4, 6, 0, 1, -2] / степени с корр = [1, 2, 0, 1, -2]

def antidervative(coef_of_derivative, const):
    # 1
    coef_of_derivative = np.append(coef_of_derivative, const)
    
    # 2
    powers = np.arange(len(coef_of_derivative))[::-1]
    
    # 3
    powers[-1] = 1
    
    # 4
    return coef_of_derivative / powers
```

```python
antidervative(np.array([4, 6, 0, 1]), -2)
```

```python
# 7 
# Что происходит
# [1, 2, 0, 1, -2]
# -> [4, 6, 0, 1]

# 1. Убрать константу: [1, 2, 0, 1]
# 2. Степени: [3, 2, 1, 0]
# 3. Корректировка: [4, 3, 2, 1]
# 4. Умножаем [1, 2, 0, 1] * скорр степени

def derivative(coef, point):
    # 1
    coef = coef[:-1]
    # 2
    powers = np.arange(len(coef))[::-1]
    # 3
    # powers = powers + 1
    powers += 1
    # 4
    derivative_coef = powers * coef
    
    return derivative_coef, np.sum(derivative_coef * point ** (powers - 1))
```

```python
derivative(np.array([1, 2, 0, 1, -2]), 1)
```

```python
derivative(np.array([0, 0, 1, 0, 0]), 1)
```
